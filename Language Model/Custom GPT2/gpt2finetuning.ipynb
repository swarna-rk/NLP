```python
import os
import re

from google.colab import drive
drive.mount('/content/drive')

#with open("merged_books.txt",'r',encoding ='utf-8') as f:
with open('/content/drive/My Drive/harrypotter1.txt', 'r') as f: 
    string = f.read()
    #print(string)
```

```python
def clean_text(string):
    pattern = '(page|PAGE|Page)(\s+\|\s+)([0-9]+)(.*)$'
    output_cleaned = re.sub('\s$', '', string, flags=re.MULTILINE)
    p=re.compile(pattern,re.MULTILINE)
    output_cleaned = p.sub(" ",output_cleaned)
    return output_cleaned

check = clean_text(string)
print(check)

with open("cleanhp_book.txt",'w',encoding='utf-8') as f:
     f.writelines(check)
    
train_doc =[]
val_doc =[]
with open("cleanhp_book.txt",'r',encoding = 'utf-8') as f:
    file_input=f.readlines()

for cnt, line in enumerate(file_input):
#         print(cnt)
        if cnt <= len(file_input)*0.90:
            train_doc.append(line)
        else:
            val_doc.append(line)

f = open('train_harry.txt', "w+",encoding = 'utf-8')
count = 0
for line in train_doc:
    count=count+1
    f.write(str(line))
    f.write("\n")
    
f.close()
print(count)

f = open('val_harry.txt', "w+",encoding = 'utf-8')
count = 0
for line in val_doc:
    count=count+1
    f.write(str(line))
    f.write("\n")
    
f.close()
print(count)
```
