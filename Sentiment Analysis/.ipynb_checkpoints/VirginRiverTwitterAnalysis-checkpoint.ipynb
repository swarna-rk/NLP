# Code to fetch tweets based on search term and date

```python
import os
import datetime as dt
import pandas as pd
import twint

def twint_search(search_term, since, until, save_path):    
    c = twint.Config()
    c.Lang = 'en'
    c.Search = "virginriver"
    c.Since = since.strftime('%Y-%m-%d %H:%M:%S')
    c.Until = until.strftime('%Y-%m-%d %H:%M:%S')
    c.Limit = 200
    # c.Hide_output = True
    c.Store_csv = True
    c.Output = save_path

    twint.run.Search(c)

def twint_search_loop(search_term, start_date, end_date, save_dir):
    try:
        os.makedirs(os.path.join(os.getcwd(),save_dir,search_term))
        print(f'Successfully created the directory {os.path.join(os.getcwd(),save_dir,search_term)}')
    except FileExistsError:
        print(f'Directory {os.path.join(os.getcwd(),save_dir,search_term)} already exists')
    
    date_range = pd.date_range(start_date, end_date)
    
    for single_date in date_range:
        since = single_date
        until = single_date + dt.timedelta(days=2)        
        save_path = os.path.join(save_dir, search_term, f'{single_date:%Y%m%d}.csv')
        print(f"Searching for tweets containing '{search_term}' from {single_date:%Y-%m-%d} and saving into {save_path}")
        twint_search(search_term, since, until, save_path)


start_date = dt.datetime(2021, 8, 1)
end_date = dt.datetime(2021, 8, 31)
save_dir = 'E:\Downloads2'
search_term = 'virginriver'
twint_search_loop(search_term,start_date, end_date,save_dir)
```

# Fetch tweets from saved csv files

```python
from glob import glob
csv_files = glob(os.path.join(save_dir, search_term, '*.csv'))

# create DataFrames for each CSV file and combine into a single df
dfs = [pd.read_csv(csv_file) for csv_file in csv_files]
tweets_df = pd.concat(dfs).reset_index(drop=True)
tweets_df.shape
```

# Import VaderSentiment and filter english tweets

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from langdetect import detect 

def detector(x):
    try:
       return detect(x)
    except:
        None 

tweets_df['lang'] = tweets_df['tweet'].apply(lambda x:detector(x))
tweets_df = tweets_df[tweets_df['lang'] == 'en']
tweets_df.shape
```

# Get PolarityScores for tweets

```python
analyzer = SentimentIntensityAnalyzer()
sentiment= tweets_df['tweet'].apply(lambda x: analyzer.polarity_scores(x))
tweets_df = pd.concat([tweets_df, sentiment.apply(pd.Series)],1)
tweets_df.drop_duplicates(subset = 'tweet',inplace = True)
tweets_df.shape
```

```python
conditions = [
    (pd.to_datetime(tweets_df['date']) < dt.datetime(2020,2,1)),
    ((pd.to_datetime(tweets_df['date']) > dt.datetime(2020,11,25)) & (pd.to_datetime(tweets_df['date']) < dt.datetime(2021,1,31))),
    ((pd.to_datetime(tweets_df['date']) > dt.datetime(2021,7,8)) & (pd.to_datetime(tweets_df['date']) < dt.datetime(2021,9,1)))    
    ]

# create a list of the values we want to assign for each condition
values = ['Season 1', 'Season 2', 'Season 3']

# create a new column and use np.select to assign values to it using our lists as arguments
tweets_df['Season'] = np.select(conditions, values)
```

# Clean the Tweets

```python
import re
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

def clean_tweet(tweet):
    temp = tweet.lower()
    temp = re.sub("'", "", temp) # to avoid removing contractions in english
    temp = re.sub("@[A-Za-z0-9_]+","", temp)
    temp = re.sub("#[A-Za-z0-9_]+","", temp)
    temp = re.sub(r'http\S+', '', temp)
    temp = re.sub('[()!?]', ' ', temp)
    temp = re.sub('\[.*?\]',' ', temp)
    temp = re.sub("[^a-z0-9]"," ", temp)
    temp = temp.split()
    temp = [w for w in temp if not w in STOPWORDS]
    temp = " ".join(word for word in temp)
    return temp

tweets_df['CleanedTweets'] =  tweets_df['tweet'] .apply(clean_tweet)
```

# Word Cloud Generator

```python
import matplotlib.pyplot as plt
from PIL import Image
from wordcloud import WordCloud

# WordCloud
stopwords = set(STOPWORDS)
stopwords.update(["netflix","series","season","episode","episodes","s","u","t","m","amp"])
wc = WordCloud(stopwords=stopwords,mode='RGBA',
             background_color='rgba(255, 255, 255, 0)')
wc.generate(' '.join(tweets_df['CleanedTweets']))


# Image process
image = Image.fromarray(wc.to_array())
background = Image.open('C:/Users/viswa/Downloads/Rectangle 5.png')
background = background.resize(image.size)
new_image = Image.alpha_composite(background, image)

# Save
plt.figure()
plt.axis('off')
fig = plt.imshow(new_image, interpolation='nearest')
plt.savefig('test1.png',
            pad_inches=0,
            format='png',
            dpi=300)
```
